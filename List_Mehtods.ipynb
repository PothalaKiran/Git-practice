List Methods

```python
#Append method -->To add the values in the end of the list
fruits = ['apple','banana','cherry']
print(fruits)
fruits.append('orange')
print(fruits)
```

```python
# Insert Method --> To add the values at specific index in the list
fruits = ['apple','banana','cherry']
print(fruits)
fruits.insert(1,'grapes')
print(fruits)
```

```python
# Reomve Method --> To reomove the specific value in the first occurence in the list

Numbers =[1,2,3,4,5,6,2,4,1]
print(Numbers)
Numbers.remove(2)
print(Numbers)
```

```python
# POP --> To reove the value at specific index

Numbers =[1,2,3,4,5,6,2,4,1]
print(Numbers)
Numbers.pop(3)
print(Numbers)
```

```python
# Index --> To see the index of specific value in the list

list1=[74,65,21,41,52,36]
print(list1)
print(list1.index(52))
```

```python
# Count --> To count the value (How many times repeated)) in the list

Numbers =[1,2,3,4,5,6,2,4,1,2,3,1]
print(Numbers)
Numbers.count(2)
```

```python
# Sort --> used to sort the values in the list

Numbers =[1,2,3,4,5,6,2,4,]
print(Numbers)
Numbers.sort()
print(Numbers)
```

```python
# Reverse --> To Reverse the values in the list
Numbers =[1,2,3,4,5,6,2,4,]
print(Numbers)
Numbers.reverse()
```

```python
# Enumerate --> Used to see the index and value

fruits = ['apple','banana','cherry']
print("Original list:", fruits)

print("\nUsing enumerate to get index and value:")
for index, value in enumerate(fruits):
    print(f"Index: {index}, Value: {value}")
```

Clear Method --> Used to clear the values in the list
