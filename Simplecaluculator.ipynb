```python
# Simple Calculator application

a= int (input("Enter first number: "))
b= int (input("Enter second number: "))
operation=input("Enter the operation: +,-,*,/ ")

if operation == "+":
    print(a+b)
if operation == "-":
    print(a-b)
if operation == "*":
    print(a*b)
if operation =="/":
  print(a/b)
```
