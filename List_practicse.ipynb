List slicing

```python
# List slicing helps us to retrive specific data points from the list
listf[2:5]
# list [Satrting values: last value +1]
```

Printing the values in the list using loops

```python
listf =["mango","apple","straberry","guvava","banana","pineapple"]
for i in listf:
    print(i)
```

List are mutable:we can change the values in the list using index

```python
#before upadting the value
print(listf[1])
#after updating the value
listf[1]="lichi"
print(listf[1])
```
